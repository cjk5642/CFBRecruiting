```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tm = pd.read_csv('total_merged.csv')
```

```python
temp1 = tm[(tm.ranking == 1) & (tm.type == 'HighSchool')].sort_values('year')
temp1 = temp1.iloc[:,12 :-12].reset_index(drop = True)
years = list(range(2000, 2019))

fig, ax = plt.subplots(figsize = (12, 8))
for j, i in zip(years, range(len(temp1))):
    a = list(map(list, zip(*list(enumerate(list(temp1.iloc[i, :]))))))
    plt.plot(a[0], a[1], label = str(j))

plt.title('Top Recruit and Outcome of Team')
plt.legend()
plt.ylabel('Rank')
plt.xlabel('Week')
plt.show()
```

```python
temp2 = tm[tm.ranking == 1]
t2 = temp2.value_counts('college').head(10)
col = list(t2.index)
coun = list(t2)

fig, ax = plt.subplots(figsize = (12, 8))
plt.bar(col, coun)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
